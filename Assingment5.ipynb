```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel('assign4.xls')
```

```python
data
```

```python
maxX=np.max(data['X'])
maxY=np.max(data['Y'])
minX=np.min(data['X'])
minY=np.min(data['Y'])
```

```python
normalized_x=[]
for i in range(data.shape[0]):
    normalized_x.append((data['X'][i]-minX)/(maxX-minX))
```

```python
normalized_y=[]
for i in range(data.shape[0]):
    normalized_y.append((data['Y'][i]-minY)/(maxY-minY))
```

```python
data['X']=normalized_x
data['Y']=normalized_y
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
x=data['X']
y=data['Y']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=10)
```

```python
def gradient(m,c,x,y):
    m_deriv=-1*(y-m*x-c)*x
    c_deriv=-1*(y-m*x-c)
    return m_deriv,c_deriv

import random
m=np.random.uniform(-1,1)
c=np.random.uniform(-1,1)
eta=0.01
max_iter=500

for iter in range(max_iter):
    for sample in range(x_train.shape[0]):
        delm,delc=gradient(m, c, x_train.iloc[sample], y_train.iloc[sample])
        m=m-eta*delm
        c=c-eta*delc
print('model parameter m and c',m,c)
```

```python
y_pred_train = []
for i in x_train:
    y_pred = (m * i) + c
    y_pred_train.append(y_pred)
y_pred_train = np.array(y_pred_train)
```

```python
y_pred_test = []
for i in x_test:
    y_pred = (m * i) + c
    y_pred_test.append(y_pred)
y_pred_test = np.array(y_pred_test)
```

```python
y_train_denorm=[]
y_train_pred_denorm=[]
for i in range(y_train.shape[0]):
    y_train_denorm.append(y_train.iloc[i]*(maxY-minY)+minY)
    y_train_pred_denorm.append(y_pred_train[i]*(maxY-minY)+minY)
y_pred_train=y_train_pred_denorm
y_train=y_train_denorm
```

```python
y_test_denorm=[]
y_test_pred_denorm=[]
for i in range(y_test.shape[0]):
    y_test_denorm.append(y_test.iloc[i]*(maxY-minY)+minY)
    y_test_pred_denorm.append(y_pred_test[i]*(maxY-minY)+minY)
y_pred_test=y_test_pred_denorm
y_test=y_test_denorm
```

```python
import math
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error
meanSquareErr=0
root_meanSquareErr = 0
meanAbsoluteError = 0
for k in range(len(y_train)):
    # print(y_train)
    meanSquareErr += (y_train[k]-y_pred_train[k]) ** 2
    meanAbsoluteError += abs(y_train[k]-y_pred_train[k])

meanSquareErr = meanSquareErr/len(y_train)
root_meanSquareErr = math.sqrt(meanSquareErr)
meanAbsoluteError = meanAbsoluteError/len(y_train)
print('Mean square error', meanSquareErr)
print('Root mean square error', root_meanSquareErr) 
print('Mean absolute error', meanAbsoluteError)
```

```python
meanSquareErr=0
root_meanSquareErr = 0
meanAbsoluteError = 0
for k in range(len(y_test)):
    # print(y_train)
    meanSquareErr += (y_test[k]-y_pred_test[k]) ** 2
    meanAbsoluteError += abs(y_test[k]-y_pred_test[k])

meanSquareErr = meanSquareErr/len(y_test)
root_meanSquareErr = math.sqrt(meanSquareErr)
meanAbsoluteError = meanAbsoluteError/len(y_test)
print('Mean square error', meanSquareErr)
print('Root mean square error', root_meanSquareErr) 
print('Mean absolute error', meanAbsoluteError)
```
